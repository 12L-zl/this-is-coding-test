준비 중인 로봇은 2 X 1 크기의 로봇으로 '0'과 '1'로 이루어진 N X N 크기의 지도에서 2 X 1 크기인 로봇을 움직여 (N, N) 위치까지 이동할 수 있도록 프로그래밍을 하려고 한다. 로봇이 이동하는 지도는 가장 왼쪽, 상단의 좌표를 (1, 1)로 하며 지도 내에 표시된 숫자 0은 빈칸을, 1은 벽을 나타낸다. 로봇은 벽이 있는 칸 또는 지도 밖으로는 이동할 수 없다. 로봇은 처음에 좌표 **(1, 1) 위치에서 가로 방향**으로 놓여져 있는 상태로 시작하며, 앞뒤 구분 없이 움직일 수 있다.

로봇이 움질일 때는 현재 놓여 있는 상태를 유지하며 이동한다. 예를 들어, 오른쪽 한 칸 이동한다면 (1, 2), (1, 3) 두 칸을 차지하게 되며, 아래로 이동한다면 (2, 1), (2, 2) 두 칸을 차지하게 된다. 로봇이 차지하는 두 칸 중 어느 한 칸이라도 (N, N) 위치에 도착하면 된다.

로봇은 아래 그림과 같이 조건에 따라 회전이 가능하다.
- 로봇은 90도씩 회전할 수 있다.
- 로봇이 차지하는 두 칸 중, 어느 칸이든 축이 될 수 있지만, 회전하는 방향 (축이 되는 칸으로부터 대각선 방향에 있는 칸)에는 벽이 없어야 한다.
- 로봇이 한 칸 이동하거나 90도 회전하는 데 걸리는 시간은 1초이다.

'0'과 '1'로 이루어진 지도 board가 주어질 때, 로봇이 (N, N) 위치까지 이동하는 데 필요한 최소 시간을 return 하도록 solution 함수를 완성하시오.

제한사항 )
- board의 한 변의 길이는 5 이상 100 이하이다.
- board의 원소는 0 또는 1 이다.
- 로봇이 처음에 놓여 있는 칸 (1, 1), (1, 2)는 항상 0으로 주어진다.
- 로봇이 항상 목적지에 도착할 수 있는 경우만 입력으로 주어진다.

```python
from collections import deque

def get_next_pos(pos, board) :
    next_pos = []  # 반환 결과(이동 가능한 위치들)
    pos = list(pos) # 현재 위치 정보를 리스트로 변환
    
    # 로봇 1번칸 (x1, y1) 로봇 2번칸 (x2, y2)
    x1, y1, x2, y2 = pos[0][0], pos[0][1], pos[1][0], pos[1][1]
    
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    for i in range(4) :
        nx1 = x1 + dx[i]
        ny1 = y1 + dy[i]
        nx2 = x2 + dx[i]
        ny2 = y2 + dy[i]
        
        # 이동하고자 하는 두 칸이 모두 비어 있다면
        if board[nx1][ny1] == 0 and board[nx2][ny2] == 0 :
            next_pos.append({(nx1, ny1), (nx2, ny2)})
    
    '''
    0 0                 1 0
    0 0                 0 0
    회전 가능           불가능
    
    4개 중 하나라도 1이 있다면 회전 불가능
    '''
    # 현재 로봇이 가로로 놓여 있는 경우, 세로로도 회전할 수 있는지
    if x1 == x2 :
        for i in [-1, 1] :  # 위쪽, 아래쪽
            if board[x1 + i][y1] == 0 and board[x2 + i][y2] == 0 :
                next_pos.append({(x1, y1), (x1 + i, y1)}) # 왼쪽 칸 기준으로 위아래 회전
                next_pos.append({(x2, y2), (x2 + i, y2)}) # 오른쪽 칸 기준으로 위아래 회전
    # 현재 로봇이 세로로 놓여 있는 경우, 가로로도 회전할 수 있는지
    elif y1 == y2 :
        for i in [-1, 1] : # 왼쪽, 오른쪽
            if board[x1][y1 + i] == 0 and board[x2][y2 + i] == 0 :
                next_pos.append({(x1, y1), (x1, y1 + i)}) # 위쪽 칸 기준으로 왼/오른쪽 회전
                next_pos.append({(x2, y2), (x2, y2 + i)}) # 아래쪽 칸 기준으로 왼/오른쪽 회전
    
    # 현재 위치에서 이동할 수 있는 모든 위치 반환
    return next_pos
```

```python
def solution(board) :
    n = len(board)
    new_board = [[1] * (n + 2) for _ in range(n + 2)] # 맵 외곽에 벽을 둠 (패딩)
    for i in range(n) :
        for j in range(n) :
            new_board[i + 1][j + 1] = board[i][j]
            
    q = deque()
    visit = []
    pos = {(1, 1), (1, 2)} # 로봇 시작 위치
    q.append((pos, 0))  # 큐에 삽입한 뒤
    visit.append(pos) # 방문 처리
    
    while q :
        pos, time = q.popleft()
        
        if (n, n) in pos :
            return time
        
        # 현재 위치에서 이동할 수 있는 위치 확인
        for next_pos in get_next_pos(pos, new_board) :
            # 아직 방문하지 않은 위치라면 큐에 삽입하고 방문 처리
            if next_pos not in visit :
                q.append((next_pos, time + 1))
                visit.append(next_pos)
    return 0
```

```python
n = int(input())
board = []
for _ in range(n) :
    board.append(list(map(int, input().split())))
```

```python
solution(board)
```
