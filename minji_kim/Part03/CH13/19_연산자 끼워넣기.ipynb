N개의 수로 이루어진 수열 A1, A2, ... AN이 주어진다. 또, 수와 수 사이에 끼워 넣을 수 있는 N-1개의 연산자가 주어진다. 연산자는 덧셈, 뺄셈, 곱셈, 나눗셈으로만 이루어져있다.

수와 수 사이에 연산자를 하나씩 넣어서, 수식을 하나 만들 수 있는데 이때 주어진 수의 순서를 바꾸면 안된다.
