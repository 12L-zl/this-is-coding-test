N X N 크기의 시험관이 있다. 시험관은 1 X 1 크기의 칸으론 나누어지며, 특정한 위치에는 바이러스가 존재할 수 잇다. 바이러스의 종류는 1 ~ K번까지 K가지가 잇으며 모든 바이러스는 이 중 하나에 속한다.

시험관에 존재하는 모든 바이러스는 1초마다 상, 하, 좌, 우의 방향으로 증식하는데, 매초 번호가 **낮은 종류의 바이러스부터** 먼저 증식한다. 또한 증식 과정에서 특정한 칸에 이미 어떠한 바이러스가 있다면, 그곳에는 다른 바이러스가 들어갈 수 없다.

시험관의 크기와 바이러스의 위치 정보가 주어졌을 때, S초가 지난 후에 (X, Y)에 존재하는 바이러스의 종류를 출력하는 프로그램을 작성하시오.

만약 S초가 지난 후에 해당 위치에 바이러스가 존재하지 않는다면, 0을 출력한다. 이때 X와 Y는 각각 행과 열의 위치이며, 시험관의 가장 왼쪽 위는 (1, 1)에 해당한다.

예를 들어 3 X 3 크기의 시험관에 서로 다른 1번, 2번, 3번 바이러스가 각각 (1, 1), (1, 3), (3, 1)에 위치해 있을 때, 2초가 지난 뒤에 (3, 2)에 존재하는 바이러스의 종류 계산해보자.


왼쪽은 1초가 지난 뒤의 시험관 상태,  
오른쪽은 2초가 지난 뒤의 시험관 상태

2초가 지난 뒤에 (3, 2)에 존재하는 바이러스의 종류는 3번이다. 따라서 3을 출력하면 정답이다.

입력 조건 )
- 첫째 줄에 자연수 N, K가 주어지며, 각 자연수는 공백으로 구분한다. (1 <= N <= 200, 1 <= K <= 1000)
- 둘째 줄부터 N개의 줄에 걸쳐 시험관의 정보가 주어진다. 각 행은 N개의 원소로 구성되며, 해당 위치에 존재하는 바이러스의 번호가 주어지며 공백으로 구분한다. 단, 해당 위치에 바이러스가 존재하지 않는 경우 0이 주어진다. 또한 모든 바이러스의 번호는 K 이하의 자연수로만 주어진다.
- N + 2 번째 줄에는 S, X, Y가 주어지며 공백으로 구분한다. (0 <= S <= 10000, 1 <= X, Y <= N)

출력 조건 )
- S초 뒤에 (X, Y)에 존재하는 바이러스의 종류를 출력한다. 만약 S초 뒤에 해당 위치에 바이러스가 존재하지 않는다면, 0을 출력한다.

```python
n, k = map(int, input().split())
data = []
for _ in range(n) :
    data.append(list(map(int, input().split())))
s, x, y = map(int, input().split())
```

```python
from collections import deque

graph = [] # 바이러스 정보
for i in range(n) :
    for j in range(n) :
        if data[i][j] != 0 : # 바이러스 존재하면
            graph.append((data[i][j], 0, i, j))  # 바이러스 종류, 시간, X, Y

graph.sort() # 낮은 번호의 종류 먼저 증식
q = deque(graph) # (1, 0, 0, 0), (2, 0, 0, 2), (3, 0, 2, 0)

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]


while q :
    virus, time, row, col = q.popleft()
    if time == s :
        break
    
    for i in range(4) :
        nrow = row + dx[i]
        ncol = col + dy[i]
        
        if nrow >= 0 and nrow < n and ncol >= 0 and ncol < n :
            if data[nrow][ncol] == 0 :
                data[nrow][ncol] = virus
                q.append((virus, time+1, nrow, ncol))
```
