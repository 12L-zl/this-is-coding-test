N X N 크기의 복도가 있다. 이 복도는 1 X 1 크기의 칸으로 나누어지며 특정한 위치에는 선생님, 학생, 혹은 장애물이 있다. 현재 학생 몇 명이 수업 시간에 몰래 복도에 나왔는데, 복도에 나온 학생들이 선생님의 감시에 들키지 않는 것이 목표이다.

각 선생님은 자신의 위치에서 상, 하, 좌, 우 4가지 방향으로 감시를 진행한다. 단, 복도에 장애물이 있으면 선생님은 장애물 뒤편에 숨어있는 학생을 볼 수 없다. 또한 선생님은 아무리 멀리 있더라도 장애물로 막히기 전까지 4방향으로 학생을 모두 볼 수 있다.

선생님 T, 학생 S, 장애물 O라고 표시한다.

3 X 3 크기의 복도에서 (3, 1)에는 선생님이 존재하며 (1, 1), (2, 1), (3, 3)의 위치에는 학생이 존재한다. 그리고 (1, 2), (2, 2), (3, 2) 위치에는 장애물이 존재한다.

이때 (3, 3)의 위치에 존재하는 학생은 장애물 뒤편에 숨어 있기에 감시를 피할 수 있지만, 나머지 두 학생은 선생님에게 들킨다.

학생들은 복도의 빈칸 중에서 장애물 설치할 위치를 골라, 정확히 3개의 장애물을 설치해야 한다. 그리고 장애물을 3개 설치해서 선생님의 감시로부터 모든 학생이 피할 수 있는지 계산해야 한다. N X N 크기의 복도에서 학생과 선생님의 위치 정보가 주어졌을 때, 장애물을 정확히 3개 설치하여 모든 학생이 선생님의 감시를 피할 수 있는지 출력하는 프로그램을 작성하시오.

N = 5일 때, 선생님과 학생의 위치 정보이다.


이때 장애물 3개를 다음처럼 설치하면, 선생님의 감시로부터 모든 학생이 피할 수 있다.


입력 조건 )
- 첫째 줄에 자연수 N이 주어진다. (3 <= N <= 6)
- 둘째 줄에 N개의 줄에 걸쳐 복도의 정보가 주어진다. 각 행에서는 N개의 원소가 주어지며, 공백으로 구분한다. 해당 위치에 학생이 있다면 S, 선생님이 있다면 T, 아무것도 존재하지 않는다면 X가 주어진다. 단, 항상 빈칸의 개수는 3개 이상으로 주어진다.

출력 조건 )
- 첫째 줄에 정확히 3개의 장애물을 설치하여 모든 학생들을 감시로부터 피하도록 할 수 있는지의 여부를 출력한다. 모든 학생들을 감시로부터 피하도록 할 수 있다면 'YES', 그렇지 않다면 'NO'를 출력한다.

```python
n = int(input())
data = []
for _ in range(n) :
    data.append(list(input().split()))
```

```python
teachers = [] # 선생님 위치   # [(0, 4), (1, 0), (3, 1), (4, 2)]
spaces = [] # 빈 공간 위치 정보

for i in range(n) :
    for j in range(n) :
        if data[i][j] == 'T' :
            teachers.append((i, j))
        if data[i][j] == 'X' :
            spaces.append((i, j))
```

```python
# 특정 방향으로 감시 진행
def dfs(x, y, direction) :
    
    '''
    3 X 3
    (0, 0) (0, 1) (0, 2)
    (1, 0) (1, 1) (1, 2)
    (2, 0) (2, 1) (2, 2)
    '''
    
    # 왼쪽
    if direction == 0 :
        while y >= 0 :  # 왼쪽으로 갈 수록  (x, 0)
            if data[x][y] == 'S' :
                return True
            if data[x][y] == 'O' :
                return False
            y -= 1
    
    # 오른쪽
    if direction == 1 :
        while y < n :
            if data[x][y] == 'S' :
                return True
            if data[x][y] == 'O' :
                return False
            y += 1
    
    # 위쪽
    if direction == 2 :
        while x >= 0 :
            if data[x][y] == 'S' :
                return True
            if data[x][y] == 'O' :
                return False
            x -= 1
    
    # 아래쪽
    if direction == 3 :
        while x < n :
            if data[x][y] == 'S' :
                return True
            if data[x][y] == 'O' :
                return False
            x += 1
            
    return False # 학생 미발견
```

```python
# 장애물 설치 이후에, 한 명이라도 학생 감지되는지
def process() :
    # 모든 선생님의 위치 하나씩 확인
    for x, y in teachers :
        # 4가지 방향으로 학생 감지할 수 있는지 확인
        for i in range(4) :
            if dfs(x, y, i) :
                return True
    return False
```

```python
# 학생이 한 명도 감지되지 않도록 설치할 수 있는지
find = False

# 빈 공간에서 3개를 뽑는 모든 조합 확인
from itertools import combinations

for i in combinations(spaces, 3) :
    for x, y in i :
        data[x][y] = 'O'
    
    # 학생이 한 명도 감지되지 않는 경우
    if not process() :
        # 원하는 경우 발견
        find = True
        break
    # 설치된 장애물 다시 없애기
    for x, y in i :
        data[x][y] = 'X'
```

```python
if find :
    print('YES')
else :
    print('NO')
```
