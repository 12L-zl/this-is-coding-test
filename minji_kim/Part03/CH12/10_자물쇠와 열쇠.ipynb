잠겨있는 좌물쇠는 격자 한 칸의 크기가 1X1인 N X N크기의 정사각 격자 형태고, 특이한 모양의 열쇠는 M X M 크기인 정사각 격자 형태로 되어 있다.

자물쇠는 홈이 파여 있고, 열쇠 또한 홈과 돌기 부분이 있다. 열쇠는 회전과 이동이 가능하며 열쇠의 돌기 부분을 자물쇠의 홈 부분에 딱 맞게 채우면 자물쇠가 열리게 되는 구조이다.

자물쇠 영역을 벗어난 부분에 있는 열쇠의 홈과 돌기는 자물쇠를 여는 데 영향을 주지 않지만, 자물쇠 영역 내에선 열쇠의 돌기부분과 자물쇠의 홈 부분이 정확히 일치해야 하며 열쇠의 돌기와 자물쇠의 돌기가 만나서는 안된다. 또한 자물쇠의 모든 홈을 채워 비워있는 곳이 없어야 자물쇠를 열 수 있다.

열쇠를 나타내는 2차원 배열 key와 자물쇠를 나타내는 2차원 배열 lock이 매개변수로 주어질 때, 열쇠로 자물쇠를 열 수 있으면 true를, 열 수 없으면 false를 return하도록 solution 함수를 완성하라

**제한사항 )**
- key는 M X M (3 <= M <= 20, M은 자연수) 크기 2차원 배열이다.
- lock은 N X N (3 <= N <= 20, N은 자연수) 크기 2차원 배열이다.
- M은 항상 N이하이다,
- key와 lock의 원소는 0 또는 1로 이루어져 있다. 이때 0은 홈 부분, 1은 돌기 부분을 나타낸다.

```python
m, n = map(int, input().split()) # key, lock

def data() :
    key = []
    lock = []
    for _ in range(m) :
        key.append(list(map(int, input().split())))
    for _ in range(n) :
        lock.append(list(map(int, input().split())))
        
    return key, lock

key, lock = data()
```

```python
# 90도 회전
import numpy as np
def rotate(key) :
    result = np.zeros((m, m))
    
    for i in range(m) : # 행 길이
        for j in range(len(key[0])) : # 열 길이
            result[j][n - i - 1] = key[i][j]
    return result

# 패딩
def shift(lock) :
    nlock = np.zeros(((n*n), (n*n)))
    for i in range(n) :
        for j in range(n) :
            nlock[i+n][j+n] = lock[i][j]    
    return nlock


def solution(key, lock) :
    for rotation in range(4) :  # 네 방향
        key = rotate(key)
        nlock = shift(lock)
        
        for x in range(n*(n-1)) :
            for y in range(n*(n-1)) :
                for i in range(m) :
                    for j in range(m) :
                        nlock[x+i][y+j] += key[i][j] 
                
                if nlock[i][j] == 1 :
                    return print('true')
                for k in range(m) :
                    for r in range(m) :
                        nlock[x+k][y+r] -= key[i][j]
    
    return print('false')
```

```python
# 두 번째 풀이
m, n = map(int, input().split()) # key, lock

key = []
lock = []
for _ in range(m) :
    key.append(list(map(int, input().split())))
for _ in range(n) :
    lock.append(list(map(int, input().split())))
    

# 90도 회전
nkey = [[0] * m for _ in range(m)]
for i in range(m) :
    for j in range(m) :
        nkey[j][m - i - 1] = key[i][j]
        
        
# 새로운 자물쇠의 중앙 부분에 기존의 자물쇠 넣기
nlock = [[0] * (n*n) for _ in range(n*n)]
for i in range(n) :
    for j in range(n) :
        nlock[i+n][j+n] = lock[i][j]
        
        
# 자물쇠 홈 부분이 모두 1로 채워졌는지 확인
def check(n, nlock) :
    for i in range(len(nlock) // n, (len(nlock) // n) * (n-1)) :
        for j in range(len(nlock) // n, (len(nlock) // n) * (n-1)) :
            if nlock[i][j] != 1 :
                return False
    return True

        
# 열쇠 회전한 뒤, 해당 열쇠로 자물쇠 탐색
def solution() :
    global n
    global m
    global nlock
    global nkey
    
    for _ in range(4) : # 네 방향
        for i in range(n * (n-1)) :
            for j in range(n * (n-1)) :

                # 자물쇠에 열쇠 끼워 넣기
                for x in range(m) :
                    for y in range(m) :
                        nlock[x + i][y + j] += nkey[x][y]

                # 자물쇠의 홈부분이 모두 1로 채워졌는지 확인
                if check(n, nlock) == True :
                    return True

                # 자물쇠에서 열쇠 다시 빼기
                for k in range(m) :
                    for l in range(m) :
                        nlock[i+k][j+l] -= nkey[k][l]

        return False
```
