### 기둥과 보 설치

[프로그래머스 링크 ](https://programmers.co.kr/learn/courses/30/lessons/60061)[]

+ 기둥과 보를 이용하여 벽면 구조물을 자동으로 세우는 로봇을 개발할 계획을 가지고 있음
+ 그에 앞서 로봇의 동작을 시뮬레이션 할 수 있는 프로그램을 만들고 있음
<br>

+ 프로그램은 2차원 가상 벽면에 기둥과 보를 이용한 구조물을 설치할 수 있는데,  
  기둥과 보는 길이가 1인 선분으로 표현되며 다음과 같은 규칙을 가지고 있음
  + 기둥은 바닥 위에 있거나 보의 한쪽 끝부분 위에 있거나, 또는 다른 기둥 위에 있어야 함
  + 보는 한쪽 끝부분이 기둥 위에 있거나, 또는 양쪽 끝부분이 다른 보와 동시에 연결되어 있어야 함
<br>

+ 단, 바닥은 벽면의 맨 아래 지면을 말함
<br>

+ 2차원 벽면은 nxn 크기 정사각 격자 형태이며, 각 격자는 1x1 크기
+ 맨 처음 벽면은 비어 있는 상태
+ 기둥과 보는 격자 선의 교차점의 걸치지 않고, 격자 칸의 각 변에 정확히 일치하도록 설치 가능

**예시**  

1. (1,0)에서 위쪽으로 기둥을 하나 설치 후, (1,1)에서 오른쪽으로 보를 하나 만듦
2. (2,1)에서 위쪽으로 기둥을 하나 설치 후, (2,2)에서 오른쪽으로 보를 하나 만듦
3. (5,0)에서 위쪽으로 기둥을 하나 설치 후, (5,1)에서 위쪽으로 기둥을 하나 더 설치함
4. (4,2)에서 오른쪽으로 보를 설치 후, (3,2)에서 오른쪽으로 보를 설치함
<br>

만약 (4,2)에서 오른쪽으로 보를 설치하지 않고, (3,2)에서 오른쪽으로 보를 설치하려 한다면 2번 규칙에 맞지 않으므로 설치 불가  
<br>

+ 기둥과 보를 삭제하는 기능도 있는데 기둥과 보를 삭제한 후에 남은 기둥과 보 또한 위 규칙을 만족해야 함
+ 만약, 작업을 수행한 결과가 조건을 만족하지 않는다면 해당 작업은 무시됨
<br>

+ 벽면의 크기 n, 기둥과 보를 설치하거나 삭제하는 작업이 순서대로 담긴 2차원 배열 build_frame이 매개변수로 주어질 때,  
  모든 명령어를 수행한 후 구조물의 상태를 return 하도록 solution 함수를 완성하기

**제한 사항**  
+ n은 5이상 100이하 자연수
+ build_frame의 세로(행) 길이는 1이상 1000 이하
+ build_frame의 가로(열) 길이는 4
+ build_frame의 원소는 [x,y,a,b] 형태
  + x,y는 기둥, 보를 설치 또는 삭제할 교차점의 좌표이며, [가로 좌표, 세로 좌표] 형태
  + a는 설치 또는 삭제할 구조물의 종류를 나타내며, 0은 기둥, 1은 보를 나타냄
  + b는 구조물을 설치할지, 혹은 삭제할지를 나타내며 0은 삭제, 1은 설치를 나타냄
  + 벽면을 벗어나게 기둥, 보를 설치하는 경우는 없음
  + 바닥에 보를 설치하는 경우는 없음
+ 구조물은 교차점 좌표를 기준으로 보는 오른쪽, 기둥은 위쪽 방향으로 설치 또는 삭제하기
+ 구조물이 겹치도록 설치하는 경우와 없는 구조물을 삭제하는 경우는 입력으로 주어지지 않음
<br>

+ 최종 구조물의 상태는 아래 규칙에 맞춰 return하기
  + return 하는 배열은 가로(열) 길이가 3인 2차원 배열로, 각 구조물의 좌표를 담고 있어야 함
  + return 하는 배열의 원소는 [x,y,a] 형식
  + x,y는 기둥, 보의 교차점 좌표이며 [가로 좌표, 세로 좌표] 형태
  + 기둥, 보는 교차점 좌표를 기준으로 오른쪽, 또는 위쪽 방향으로 설치되어 있음을 나타냄
  + a는 구조물의 종류를 나타내며, 0은 기둥, 1은 보를 나타냄
  + return 하는 배열은 x좌표를 기준으로 오름차순 정렬하며, x좌표가 같을 경우 y좌표 기준으로 오름차순 정렬
  + x,y 좌표가 모두 같은 경우 기둥이 보보다 앞에 오면 됨

```python
def possible(answer):
    for x, y, ki_or_bo in answer:
        if ki_or_bo == 0: # 기둥을 설치한다고 할때
            # 바닥이거나 기둥 출발점이거나, 보의 한쪽 끝부분 위라면 가능 
            if y == 0 or [x-1, y, 1] in answer or [x, y, 1] in answer or [x, y-1, 0] in answer:
                continue
            else:
                return False
        elif ki_or_bo == 1: # 보를 설치한다고 할때
            # 기둥 출발점이거나, 보를 설치했을때 기둥이랑 연결되거나, 왼쪽점과 오른쪽점에 보가 세워졌으면 가능
            if [x, y-1, 0] in answer or [x+1, y-1, 0] in answer or ([x-1, y, 1] in answer and [x+1, y, 1] in answer):
                continue
            else:
                return False
    return True
```

```python
def solution(n, build_frame):
    answer = []
    for i in range(len(build_frame)):
        x, y, ki_or_bo, plus_or_minus = build_frame[i]
        if plus_or_minus == 0: # 삭제라면
            answer.remove([x, y, ki_or_bo]) # 일단 삭제해보고
            if possible(answer) == False: # 불가능하다고 나오면
                answer.append([x, y, ki_or_bo]) # 다시 추가해주기
        if plus_or_minus == 1: # 추가라면
            answer.append([x, y, ki_or_bo]) # 일단 추가해보고
            if possible(answer) == False: # 불가능하다고 나오면
                answer.remove([x, y, ki_or_bo]) # 다시 삭제해주기
    return sorted(answer)
```

```python
solution(5,[[1,0,0,1],[1,1,1,1],[2,1,0,1],[2,2,1,1],[5,0,0,1],[5,1,0,1],[4,2,1,1],[3,2,1,1]])
```
